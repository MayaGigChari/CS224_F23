```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Read the TSV file in Python
data = pd.read_csv('q4.data', sep='\t')

# Select columns that start with 'rs'
snps = data.filter(regex=r'^rs')
```

```python
# Set random seed for reproducibility
np.random.seed(0)

# Separate the SNP data and population labels
y = data['population']

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(snps)

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Concatenate the population labels with principal components
final_df = pd.concat([principal_df, y], axis=1)

# Define colors and labels for populations
colors = {'EUR': 'red', 'AMR': 'blue', 'AFR': 'green', 'ASN': 'orange'}  # Update with your population labels
labels = list(colors.keys())

# Plot PC1 against PC2
plt.figure(figsize=(10, 8))
for label, color in zip(labels, colors.values()):
    subset = final_df[final_df['population'] == label]
    plt.scatter(subset['PC1'], subset['PC2'], c=color, label=label, alpha=0.7)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot of Genetic Data')
plt.legend()
plt.show()
```

```python
# Perform KMeans clustering
kmeans = KMeans(n_clusters=4, n_init=5, random_state=0)  # Use n_init=5 for multiple initializations
cluster_labels = kmeans.fit_predict(snps)

# Assign cluster labels to the principal components DataFrame
principal_df['Cluster'] = cluster_labels + 1  # Adjusting labels to start from 1

# Count the number of individuals in each cluster
cluster_sizes = principal_df['Cluster'].value_counts().sort_values(ascending=False)

# Rename clusters based on size (Cluster 1 -> largest, Cluster 4 -> smallest)
rename_dict = {cluster: f'Cluster{idx+1}' for idx, cluster in enumerate(cluster_sizes.index)}
principal_df['Cluster'] = principal_df['Cluster'].map(rename_dict)

# Plot PC1 against PC2, colored by cluster assignments
plt.figure(figsize=(10, 8))
for cluster in principal_df['Cluster'].unique():
    subset = principal_df[principal_df['Cluster'] == cluster]
    plt.scatter(subset['PC1'], subset['PC2'], label=cluster, alpha=0.7)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot of Genetic Data with KMeans Clustering')
plt.legend()
plt.show()
```

```python
cluster_to_population = {
    'Cluster1': 'EUR',
    'Cluster2': 'ASN',
    'Cluster3': 'AFR',
    'Cluster4': 'AMR'
}

# Map cluster labels to population labels in the DataFrame
principal_df['Assigned Population'] = principal_df['Cluster'].map(cluster_to_population)

# Assuming 'population' column exists in the original data DataFrame
correct_assignments = sum(data['population'] == principal_df['Assigned Population'])
total_assignments = len(principal_df)
fraction_agreement = correct_assignments / total_assignments

print(f'Fraction of cluster assignments agreeing with true population labels: {fraction_agreement:.3f}')
```

```python
# Calculate fraction agreement for each population
fractions_agreement_per_population = {}
for population in cluster_to_population.values():
    subset_df = principal_df[principal_df['Assigned Population'] == population]
    correct_assignments = sum(data[data['population'] == population].index.isin(subset_df.index))
    total_assignments = len(subset_df)
    fraction_agreement = correct_assignments / total_assignments
    fractions_agreement_per_population[population] = fraction_agreement

# Print fraction agreement for each population
print("\nFraction agreement for each population:")
for population, fraction in fractions_agreement_per_population.items():
    print(f'{population}: {fraction:.3f}')
```
