```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd

import statsmodels.api as sm

import statsmodels.formula.api as smf

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

#ridge regression problem
```

```python
ridge_train_geno = np.loadtxt("ridge.training.geno",delimiter = " ")
ridge_test_geno = np.loadtxt("ridge.test.geno",delimiter = " ")
ridge_train_pheno = np.loadtxt("ridge.training.pheno",delimiter = " ")
ridge_test_pheno = np.loadtxt("ridge.test.pheno",delimiter = " ")
```

```python
lmbdas = [0.001,2,5,8]
mses = []
mses_train= []
for i in lmbdas:
    mod_ridge = Ridge(alpha=i)
    mod_ridge.fit(ridge_train_geno, ridge_train_pheno)

    ridge_predict_pheno = mod_ridge.predict(ridge_test_geno)
# Step 3: Calculate the Mean Squared Error
    mses.append(mean_squared_error(ridge_test_pheno, ridge_predict_pheno))
    ridge_train_predict = mod_ridge.predict(ridge_train_geno)
    mses_train.append(mean_squared_error(ridge_train_pheno, ridge_train_predict))

    
#as the value of lambda increases, the mean squared error decreases. 

```

```python
plt.plot(lmbdas, mses)
plt.title('MSE on test data across lambdas')  # Add a title

```

```python
plt.plot(lmbdas,mses_train)
plt.title('MSE on train data across lambdas')  # Add a title


#as we can see the training error is very low in 
```
